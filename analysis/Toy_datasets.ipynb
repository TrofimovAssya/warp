```python
%matplotlib inline
from sklearn import datasets
import numpy
from matplotlib import pyplot as plt
```

```python
noisy_circles = datasets.make_circles(n_samples=500, factor=.5,
                                      noise=.05)
```

```python
plt.scatter(noisy_circles[0][:,0], noisy_circles[0][:,1])
```

```python
noisy_moons = datasets.make_moons(n_samples=500, noise=.05)
noisy_moons
```

```python
plt.scatter(noisy_moons[0][:,0], noisy_moons[0][:,1])
```
