```python
%matplotlib inline
import numpy
from matplotlib import pyplot as plt
```

```python
#### generating one class

ax = numpy.random.normal(3, 0.5, (100))
ay = numpy.random.normal(3, 0.5, (100))
theta = 35
axneg = ax*numpy.cos(theta)-ay*numpy.sin(theta)
ayneg = ax*numpy.sin(theta)+ay*numpy.cos(theta)
x1 = numpy.concatenate((ax,axneg))
y1 = numpy.concatenate((ay,ayneg))

#### generating the other class

bx = numpy.random.normal(3, 0.5, (100))
by = numpy.random.normal(-3, 0.5, (100))
theta = 35
bxneg = bx*numpy.cos(theta)-by*numpy.sin(theta)
byneg = bx*numpy.sin(theta)+by*numpy.cos(theta)
x2 = numpy.concatenate((bx,bxneg))
y2 = numpy.concatenate((by,byneg))

x = numpy.concatenate((x1,x2))
y = numpy.concatenate((y1,y2))
```

```python
plt.scatter(x,y)
```

```python
####### other way of generating less ugly data
#### generating one class

ax = numpy.random.normal(3, 0.5, (100))
ay = numpy.random.normal(3, 0.5, (100))
theta = 35
axneg = numpy.random.normal(-3, 0.5, (100))
ayneg = numpy.random.normal(-3, 0.5, (100))
x1 = numpy.concatenate((ax,axneg))
y1 = numpy.concatenate((ay,ayneg))

#### generating the other class

bx = numpy.random.normal(3, 0.5, (100))
by = numpy.random.normal(-3, 0.5, (100))
theta = 35
bxneg = numpy.random.normal(-3, 0.5, (100))
byneg = numpy.random.normal(3, 0.5, (100))
x2 = numpy.concatenate((bx,bxneg))
y2 = numpy.concatenate((by,byneg))

x = numpy.concatenate((x1,x2))
y = numpy.concatenate((y1,y2))
```

```python
plt.scatter(x,y)
```
